```python
x=[[1,148,24,1300],
  [1,134,25,1289],
  [1,354,43,1156],
  [1,149,24,1600],
  [1,234,25,1289],
  [1,354,43,1156],
  [1,148,34,1389],
  [1,134,25,1289],
  [1,392,43,1156]]
x=np.array(x)
one=np.ones(x.shape[0])
one
x
```

```python
y=[1334,123,432,4444,222,567,3566,22456,4566]
y
```

```python
xTx=x.T.dot(x)
```

```python
xTx_inv=np.linalg.inv(xTx)
```

```python
w_full=xTx_inv.dot(x.T).dot(y)
```

```python
w0=w_full[0]
w=w_full[1:]
w0,w
```

```python
def train_linear_regression(x,y):
    ones=np.ones(x.shape[0])
    x=np.column_stack([ones,x])
    xTx=x.T.dot(x)
    xTx_inv=np.linalg.inv(xTx)
    w_full=xTx_inv.dot(x.T).dot(y)
    return w_full[0],w_full[1:]
```
